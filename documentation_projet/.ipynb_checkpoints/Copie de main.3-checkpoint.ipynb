```python
Lmax = 300e-3 #300 mm = 0,300 m 
b = 29e-3     #29 mm = 0,029 m 
d = 1e-3     #1  mm = 0,001 m

Igz = (b*d**3)/12
E = 69e9    #Module d'young (69 * 10^9 Pa = 69 * 10^9 N/m²)

poid1= 5e-3                    # 5 g = 0,005 Kg
poid2= 50e-3                   # 50 g = 0,05 Kg
poid3= (50+20+20+10+5)*0.001   # 105 g = 0,105 Kg

F1 = poid1*9.81   # Kg * N/Kg = N
F2 = poid2*9.81   # Kg * N/Kg = N
F3 = poid3*9.81   # Kg * N/Kg = N



def y1(L):
    return ((6*F1*L)/(b*d**2))/E    #  ((F*L**3)/(6*E*Igz))*((x**3)/(L**3) - 3*((x**2)/(L**2)))
def y2(L):
    return ((6*F2*L)/(b*d**2))/E    #  ((F*L**3)/(6*E*Igz))*((x**3)/(L**3) - 3*((x**2)/(L**2)))
def y3(L):
    return ((6*F3*L)/(b*d**2))/E    #  ((F*L**3)/(6*E*Igz))*((x**3)/(L**3) - 3*((x**2)/(L**2)))

def ε(y):
    return y*2
```

```python
import matplotlib.pyplot as plt
import numpy as np
x_range = []
x1_range = []

Y_range = []
y1_range = []
y2_range = []
y3_range = []

εP = []
ε1 =[]
ε2 =[]
ε3 =[]

for i in np.arange(0,Lmax,1e-4):
    x_range.append(i)
    
    y1_range.append(y1(i)*10**2)
    y2_range.append(y2(i)*10**2)
    y3_range.append(y3(i)*10**2)
    
     
    ε1.append(ε(y1(i)*10**2))
    ε2.append(ε(y2(i)*10**2))
    ε3.append(ε(y3(i)*10**2))
    


plt.plot(x_range,y1_range,'m',label = "5g")
plt.plot(x_range,y2_range,'r',label = "50g")
plt.plot(x_range,y3_range,'b',label = "105g")

plt.title("Écart relatif de la longueur en fonction d'une distance")
plt.xlabel("Distance du poids en m")
plt.ylabel("Écart relatif en %")
plt.legend()
plt.show()
resultat1 = (y1(0.3) * 300)
print("pour un poids de   5g posé à 30 cm on a ",round(resultat1,3),"mm de déformation")
resultat2 = (y2(0.3) * 300)
print("pour un poids de  50g posé à 30 cm on a ",round(resultat2,3),"mm de déformation")
resultat3 = (y3(0.3) * 300)
print("pour un poids de 105g posé à 30 cm on a ",round(resultat3,3),"mm de déformation")
plt.figure()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Lp = 0.3       #300 mm = 0,300 m 
bp = 0.029     #29 mm = 0,029 m 
dp = 0.001     #1  mm = 0,001 m

Igz = (bp*dp**3)/12
E = 69e9    #Module d'young (69 * 10^9 Pa = 69 * 10^9 N/m²)



x1_range = []
Y_range = []
εP = []

def Y(P):
    return ((6*P*9.81*Lp)/(bp*dp**2))/E
def ε(y):
    return y*2

for i in np.arange(0,105,5):
    x1_range.append(i)
    Y_range.append(Y(i)/10)
    εP.append(ε(Y(i)/10))
#plt.plot(x1_range,Y_range)
plt.plot(x1_range,εP)
```
